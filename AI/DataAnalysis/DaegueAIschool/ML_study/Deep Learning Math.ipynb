```python
def linear_function(x):
    a = 0.5
    b = 2

    return a * x + b
```

```python
x = np.arange(-5,5,0.1)
y = linear_function(x)
plt.plot(x,y) # 선형 회귀 그래프 (1차 함수)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

```python
# 2차 함수
def quadratic_function(x):
    a = 1 #(기울기)
    b = -2 #(졀편)
    c = -2
    
    return a*x**2+b*x+c
```

```python
x = np.arange(-5,5,0.1)
y = quadratic_function(x)
plt.plot(x,y) # 선형 회귀 그래프 (2차 함수)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic function')
```

```python
# 3차 함수
def cubic_function(x):
    a = 4 #기울기
    b = 0  #절편
    c = -1
    d = -8

    return a*x**3+b*x**2+c*x+d
```

```python
x = np.arange(-5,5,0.1)
y = cubic_function(x)
plt.plot(x,y) # 선형 회귀 그래프 (3차 함수)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic function')
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def AND(x1, x2):
  input = np.array([x1, x2])
  weights = np.array([0.4,0.4])
  bias = -0.6

  value = np.sum(input * weights) + bias

  if value <= 0:
      return 0
  else:
      return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
x1 = np.arange(-2, 2, 0.01)
x2 = np.arange(-2, 2, 0.01)
bias = -0.6

y = (-0.4 * x1 - bias) / 0.4

plt.axvline(x=0)
plt.axhline(y=0)

plt.plot(x1, y, 'r--')
plt.scatter(0,0,color='black',marker='^',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(1,1,color='orange',marker='o',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
```

```python
def OR(x1, x2):
  input = np.array([x1, x2])
  weights = np.array([0.4,0.4])
  bias = -0.3

  value = np.sum(input * weights) + bias

  if value <= 0:
      return 0
  else:
      return 1
```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
x1 = np.arange(-2, 2, 0.01)
x2 = np.arange(-2, 2, 0.01)
bias = -0.3

y = (-0.4 * x1 - bias) / 0.4

plt.axvline(x=0)
plt.axhline(y=0)

plt.plot(x1, y, 'r--')
plt.scatter(0,0,color='black',marker='^',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(1,1,color='orange',marker='o',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
```

```python
def NAND(x1, x2):
  input = np.array([x1, x2])
  weights = np.array([-0.6,-0.6])
  bias = 0.7

  value = np.sum(input * weights) + bias

  if value <= 0:
    return 0
  else:
    return 1
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

```python
def XOR(x1, x2):
    s1 =NAND(x1,x2)
    s2 = OR(x1,x2)
    y = AND(s1,s2)
    return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```
